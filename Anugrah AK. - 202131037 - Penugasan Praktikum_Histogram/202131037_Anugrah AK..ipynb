### Konvolusi

```python
#202131037_Anugrah AK.
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
#202131037_Anugrah AK.
image1 = imread(fname="citra_medis1.jpg")
print(image1.shape)
plt.imshow(image1,cmap='gray')
```

```python
kernel = np.array([[-1, 0, -1],
                   [0, 4.3, 0],
                  [-1, 0, -1]])
imageOutput = cv2.filter2D(image1, -1, kernel)

fig, axes = plt.subplots(1,2, figsize=(12,12))
ax = axes.ravel()

ax[0].imshow(image1, cmap ='gray')
ax[0].set_title("Citra Input")
ax[1].imshow(imageOutput, cmap = 'gray')
ax[1].set_title('Citra Output')
```

```python
#202131037_Anugrah AK.
img = cv2.imread("citra_medis.jpg")
```

```python
#202131037_Anugrah AK.
cv2.imshow("Example",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#202131037_Anugrah AK.
print(img.shape)
```

```python
#202131037_Anugrah AK.
[baris, kolom] = img.shape[:2]
```

### Membuat Histogram

```python
#202131037_Anugrah AK.
hist = cv2.calcHist([img], [0], None,[256],[0,256])
hist
```

```python
#202131037_Anugrah AK.
fig, axs = plt.subplots(1,2, figsize =(15,5))
axs[0].imshow(img)
axs[1].plot(hist, color = 'purple')
plt.show()
```

### Konversi BGR ke RGB

```python
#202131037_Anugrah AK.
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#202131037_Anugrah AK.
fig, axs = plt.subplots(1,2, figsize =(15,5))
axs[0].imshow(rgb)
axs[1].plot(hist, color = 'purple')
plt.show()
```

### 2 Versi Histogram

```python
#202131037_Anugrah AK.
fig, axs = plt.subplots(2,2, figsize =(22,7))

axs[0,0].imshow(img)
axs[0,1].plot(hist, color = 'purple')

axs[1,0].imshow(img)
axs[1,1].hist(img.ravel(),256, [0,256], color = 'purple')
plt.show()
```

### Histogram Abu-Abu

```python
#202131037_Anugrah AK.
abuabu = cv2.imread("citra_medis.jpg", 0)
hist2 = cv2.equalizeHist(abuabu)

fig, axs = plt.subplots(1,2, figsize =(15,5))
axs[0].imshow(cv2.cvtColor(abuabu,cv2.COLOR_BGR2RGB))
axs[1].hist(abuabu.ravel(), 256, [0,256], color= "purple")
plt.show()
```

### Meningkatkan Kecerahan

```python
#202131037_Anugrah AK.
beta = 33
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = img[x,y]+beta
        citra_cerah[x,y] = gyx
citra_cerah = citra_cerah.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize =(21,7))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,256], color="purple")

axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(), 256, [0,256], color="green")
plt.show()
```

### Meregangkan Kontras

```python
#202131037_Anugrah AK.
alpa = 1.5
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = img[x,y]*alpa
        citra_kontras[x,y] = gyx
citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize =(21,7))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,256], color="purple")

axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(), 256, [0,256], color="green")
plt.show()
```
